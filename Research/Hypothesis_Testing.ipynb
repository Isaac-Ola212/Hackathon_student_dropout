### Hypothesis 2 – Gender and Dropout

H₀: There is no significant difference in dropout rates between male and female students.

H₁: There is a significant difference in dropout rates between male and female students.

Test: Independent t-test

In the context of the dataset: The analysis will compare dropout rates between male and female students to determine if gender is associated with student dropout.

```python
male_dropout = df[df["Gender"] == 0]["Dropout"].dropna()
female_dropout = df[df["Gender"] == 1]["Dropout"].dropna()
```

```python
t_stat, p_value = stats.ttest_ind(male_dropout, female_dropout)
```

```python
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

#### What this code does

- The dataset was divided into two groups based on gender (0 = male, 1 = female).

- An independent samples t-test was performed to compare dropout rates between male and female students.

- The test outputs a t-statistic and a p-value to determine whether the difference between the two groups is statistically significant.

### Interpretation

- The test produced a t-statistic of -0.57 and a p-value of 0.569.

- Since the p-value is greater than 0.05, the null hypothesis is not rejected.

- This indicates that there is no statistically significant difference in dropout rates between male and female students.

### In Context of the Dataset

- The results suggest that gender does not significantly impact student dropout.

- Male and female students show similar dropout patterns, indicating that other factors may be more important.
