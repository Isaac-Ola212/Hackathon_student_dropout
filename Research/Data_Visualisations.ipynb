```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the cleaned dataset
df = pd.read_csv("../Dataset/Processed/clean_student_dropout_dataset.csv")
df.head()
```

```python
# Dropout Distribution
sns.countplot(x='Dropout', data=df)
plt.title('Distribution of Dropout')
plt.show()
```

* Understand overall dropout rate

The distribution shows that a significant proportion of students do not complete their studies, indicating that student dropout is a relevant issue within the dataset. This justifies further analysis to identify the key factors contributing to student attrition.

```python
# Attendance vs Dropout
sns.boxplot(x='Dropout', y='Attendance_Rate', data=df)
plt.title('Attendance Rate by Dropout Status')
plt.show()
```

* Students with lower attendance are more likely to drop out

Students who dropped out have noticeably lower attendance rates compared to those who remained enrolled or graduated. This suggests that attendance is a strong indicator of student engagement and is highly associated with dropout risk. Improving attendance could play a critical role in reducing dropout rates.

```python
# Gender Distribution
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

* Dropout rates differ slightly across gender groups

The visualisation shows some variation in dropout rates across gender groups. However, the difference is not substantial, suggesting that gender alone may not be a strong predictor of dropout. This relationship should be interpreted cautiously, as it may reflect underlying social or contextual factors rather than a direct causal effect.

```python
# Scholarship Status vs Dropout
sns.barplot(x='Scholarship', y='Dropout', data=df)
plt.title('Scholarship Status vs Dropout')
plt.show()
```

* Students without scholarships show higher dropout rates

Students without scholarships exhibit higher dropout rates compared to those receiving financial support. This indicates that financial assistance may play an important role in student retention, and providing support mechanisms could help reduce dropout rates.

```python
# Correlation Heatmap (numeric features only)
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include='number')
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

* Identify strongest relationships between variables

The correlation heatmap highlights the strength and direction of relationships between variables. Attendance shows a notable relationship with dropout, supporting earlier findings. Most other variables show weak to moderate correlations, indicating that dropout is influenced by a combination of factors rather than a single variable.

## Additional Visuals

```python
# Academic Performance vs Dropout
sns.boxplot(x='Dropout', y='GPA', data=df)
plt.title('GPA vs Dropout')
plt.show()
```

Lower grades may correlate with higher dropout

```python
# Age Distribution
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

Identify age groups with higher dropout rates

```python
# Department Analysis
sns.barplot(x='Department', y='Dropout', data=df)
plt.title('Dropout Rate by Department')
plt.xticks(rotation=45)
plt.show()
```

Certain courses may have higher dropout rates

# Conclusion

In conclusion, the analysis identified attendance rate as the strongest predictor of student dropout, with lower attendance significantly associated with higher dropout rates. Scholarship status also plays an important role, indicating that financial support contributes to student retention. While some differences were observed across gender groups, this relationship is less pronounced and should be interpreted cautiously.

These findings highlight the importance of early intervention strategies focused on improving attendance and providing financial support to reduce student dropout.
