# **ETL - Extract, Transform, and Load**

## Objectives

Prepare the student dropout dataset for analysis by loading and cleaning the data using basic pandas methods.

## Inputs

- Dataset: student_dropout_dataset_v3.csv  
- Source: https://www.kaggle.com/datasets/meharshanali/student-dropout-prediction-dataset

## Outputs

- Cleaned dataset saved as clean_student_dropout_dataset.csv  
- Data ready for statistical analysis and modelling  

## Additional Comments

Basic cleaning steps were applied, including handling missing values and removing duplicates. Categorical variables were kept in their original format as conversion was not needed at this stage.



---

# Extract 

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Dataset/Raw/student_dropout_dataset_v3.csv")
df.head()
```

---

# Transform 

```python
# Check shape of dataset
df.shape
```

```python
# Check data types and non-null values
df.info()
```

#### What the code does
- The code shows us the structure of the dataset (number of rows, columns, and data types) and shows missing values, which is important for identifying data quality issues before cleaning.

```python
# Check for missing values
df.isnull().sum()
```

#### What the code does
- Missing values are found in several columns, including Family_Income, Study_Hours_per_Day, Stress_Index, and Parental_Education which need to be removed. 

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df["Internet_Access"] = df["Internet_Access"].map({"No": 0, "Yes": 1})
df["Part_Time_Job"] = df["Part_Time_Job"].map({"No": 0, "Yes": 1})
df["Scholarship"] = df["Scholarship"].map({"No": 0, "Yes": 1})
df["Dropout"] = df["Dropout"].astype(int)
```

#### What the code does
- Categories were converted into numerical format using allowing the data to be used in statistical analysis and machine learning models.

```python
# Check again
df.isnull().sum()
```

---

# Load 

```python
df.to_csv("../Dataset/Processed/clean_student_dropout_dataset.csv", index=False)
```
